```python
from matplotlib.font_manager import fontManager
import matplotlib as mpl

fontManager.addfont("ChineseFont.ttf")
mpl.rc('font',family='ChineseFont')
```

```python
import matplotlib.pyplot as plt

student1 = {'國文':78, '英文':80, '數學':92, '歷史':75, '探索':85}
student2 = {'國文':90, '英文':75, '數學':98, '歷史':65, '探索':75}

subjects = list(student1.keys())
student1_value = list(student1.values())
student2_value = list(student2.values())

figure = plt.figure(figsize=(8, 5),dpi=150,facecolor='#eeeeee')
axes = figure.add_subplot(1,1,1)
axes.plot(subjects,student1_value,marker='o',label="徐xx")
axes.plot(subjects,student2_value,marker='o',linestyle='-.',label="張xx")
axes.set_title("張xx vs 徐xx")
axes.set_xlabel("科目")
axes.set_ylabel("分數")
axes.legend()
xlocation = axes.get_xticks()
for index in range(len(xlocations)):
    x = xlocations[index]
    student1_y = student1_value[index]
    student2_y = student2_value[index]
    # print("x座標:",x)
    # print("s1_y_座標:",student1_y)
    # print("s2_y_座標:",student2_y)
    # print("==================")
    axes.annotate(str(student1_y),(x-0.05,student1_y+1))
    if student2_y == 98:
        axes.annotate(str(student2_y),(x,student2_y),xytext=(x+0.3, student2_y+5), arrowprops=dict(facecolor='red', shrink=0.05))
    else:
        axes.annotate(str(student2_y),(x-0.05,student2_y+1))
    axes.set_xlim(-0.5,4.5)
    axes.set_ylim(60,105)
axes.grid(axis='y',color='blue',alpha=0.1,linestyle=':')
plt.show()
```
